```python
# Import necessary library
import pandas as pd
import numpy as np
```

```python
# Read the input csv
data = pd.read_csv('data.csv')
```

```python
# create a list of our conditions
conditions = [
    (data['energy'] >= 0.0)& (data['energy'] <= 0.15),
    (data['energy'] > 0.15)& (data['energy'] <= 0.30),
    (data['energy'] > 0.30)& (data['energy'] <= 0.45),
    (data['energy'] > 0.45)& (data['energy'] <= 0.60),
    (data['energy'] > 0.60)& (data['energy'] <= 0.75),
    (data['energy'] > 0.75)& (data['energy'] <= 0.90),
    (data['energy'] > 0.90)
    ]

# create a list of the values we want to assign for each condition
values = ['angry', 'disgust', 'sad', 'nervous','neutral','surprised','happy']

# create a new column and use np.select to assign values to it using our lists as arguments
data['emotion'] = np.select(conditions, values)

# display updated DataFrame
data.head()
```

```python
data.to_csv('labelled_data.csv',index = False)
```
